```python
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
```

```python
boston = pd.read_csv('data/boston.csv', sep='\t')
boston.head()
```

```python
boston.shape
```

```python
m = smf.ols('medv ~ black', data=boston).fit()
```

```python
m.params
```

The B0 or the intercept signifies where the linear regression line will meet the y axis. This number shifts the line up or down, in this case up ~5. The slope or b1 value is the relatioship between the x variable and the y variable. In this case as the average number of rooms increases by one unit, the median value of the homes increases by .05. 

```python
x = np.linspace(boston.black.min(), boston.black.max(), 10)
hatY = m.params[0] + m.params[1]*x
```

```python
_ = plt.scatter(boston.black, boston.medv, edgecolor='k')
_ = plt.plot(x, hatY, c='r')
_ = plt.xlabel('black')
_ = plt.ylabel('median house price')
```

```python
m = smf.ols('medv ~ nox', data=boston).fit()
m.params
x = np.linspace(boston.nox.min(), boston.nox.max(), 10)
hatY = m.params[0] + m.params[1]*x
_ = plt.scatter(boston.nox, boston.medv, edgecolor='k')
_ = plt.plot(x, hatY, c='r')
_ = plt.xlabel('nox')
_ = plt.ylabel('median house price')
```
